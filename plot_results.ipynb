### Galaxy MNist

```python
# function to load results from cluster

def get_results(fname, repetitions, variable, p_order_approx):
    outputs = np.zeros((len(p_order_approx),len(variable), repetitions))
    outputs = outputs.tolist()
    for i in range(repetitions):
        repId = i+1
        results = np.load(fname + str(repId) + '.npz')
        for p,key in enumerate(results.files):
            for s in range(len(variable)):
                outputs[p][s][i] = results[key][s][i]   

    output = np.mean(np.array(outputs), -1)
    return output

def plot_results(output, variable, p_order_approx, title):
    for i in range(len(p_order_approx)):
        plt.plot(variable, output[i], label='p=' + str(p_order_approx[i]), marker='s')
    plt.xlabel(title)
    plt.ylabel('Power')
    # legend outside
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylim(0,1)
```

```python
fname = './results/galaxy_results/galaxy_deep_bs_250_early_stop'  # './results/galaxy_results/galaxy_deep_new_params'
# fname =  './results/galaxy_results/galaxy_deep_new_params'
corruptions = (0.1, 0.15, 0.20, 0.25, 0.3, 0.35, 0.4)
repetitions = 200
p_order_approx = np.array([1, 2, 4, 6, 10])
output = get_results(fname, repetitions, corruptions, p_order_approx)
plot_results(output, corruptions, p_order_approx, 'Corruption level')

# save output in a npz file
# np.save('./results/galaxy_results/best_20_epochs_sigma0_25.npy', output)
```
